# video writing using OpenCV

```python
import cv2
```

```python
cap = cv2.VideoCapture(0)


fourcc = cv2.VideoWriter_fourcc(*'XVID')
# creating a instance for video writer
out = cv2.VideoWriter('newOutPut1.avi',fourcc,20.0,(640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        
        # to find the width and height of the frame
        print('width',cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print('height',cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
        # out is the instance of video writer
        # our video will save in the BGR mode
        out.write(frame)
        
        # converting the frame to gray
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        # just showing the video in gray
        cv2.imshow('frame',gray)

        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break

        
cap.release()
out.release()
cv2.destroyAllWindows()
```
