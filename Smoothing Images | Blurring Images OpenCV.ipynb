# Smoothing Images | Blurring Images

Image smoothing is a digital image processing technique that reduces and prevent image noises with the help of kernal or filters
* It is useful for removing noise
* removes high frequency content (eg: noise, edges) from the image.

in image processing a kernal, convolution matic,or mask is a small matrix . It is used for blurring ,sharpening,edge detection and more



```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Noise_salt_and_pepper.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

kernal = np.ones([5,5],np.float32)/25
dest =  cv.filter2D(img, -1, kernal)
# blur method is also called averagin method
# args    (source,kernal)
blur = cv.blur(img,(5,5)) 
# gaussian blur-the pixel located on the sides have lower weight, located center have higher weight
# args    (source,kernal,sigmavalue)
gaussian = cv.GaussianBlur(img, (5,5), 0)
# Median blur - it replace each pixels value with the median of its neighbouring pixels.mostly for salt and peper noise
# kernal size must be odd
# args    (source,kernalsize)
median = cv.medianBlur(img, 5)
# bilateral filter - it help to shaprpen the borders
# arg_2nd = diameter of each pixel neighbourgood that using the filter
# args               (source,arg_2nd,filter sigma color)
bilaterl = cv.bilateralFilter(img,9,75,75)

titles = ['image','2D convolution','blur','gaussian blur','median blur','bilaterl filter']
images = [img,dest,blur,gaussian,median,bilaterl]
size = len(images) 

for i in range(size):
    plt.subplot(2,3,i+1)

    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
