# Matplotlib with openCV


```python
import cv2 as cv
import matplotlib.pyplot as plt
```

We dont get the exact clear image using plt.show() ,because in openCV read the imagein  BGR format ,
but in matplotlib read the image in RGB format, so first we need to convert our image into RGB color space

```python
img1= cv.imread('lena.jpg')

# converting the color format BGR to RGB for show the image using plt.show() 
img2 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)

cv.imshow('image',img1)

plt.imshow(img2)
# if we need to remove the ticks
plt.xticks([]),plt.yticks([])
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```
